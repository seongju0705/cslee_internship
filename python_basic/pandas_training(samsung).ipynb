## 2. 데이터 가공

```python
# 평균, 표준편차, 사분위수 등 주요 통계 지표 확인
df.describe()
```

```python
# 'Close' 기준 최대값, 최소값, 평균값
print(df['Close'].max())
print(df['Close'].min())
print(df['Close'].mean())
```

```python
# 특정 컬럼 선택
df['Close']
```

```python
# 상위 3개의 로우 선택
df.head(3)
```

```python
# 'Open' 대비 'Close'가 높았던 날의 데이터 선택
df[df['Open'] < df['Close']]
```

```python
# 'Close' 기준 가장 주가가 높았던 날의 데이터 선택
df[df['Close'] == max(df['Close'])]
```

```python
# 'Close' 기준 주가가 90000원 이상이었던 날짜 반환
df[df['Close'] >= 90000].index
```

```python
# 2021년 1월 15일 데이터 반환
df.loc['2021-01-15']
```

```python
# 2021년 1월 11일의 High와 Low값 반환
df.loc['2021-01-11', ['High', 'Low']]
```

```python
# 100번째 데이터 선택
df.iloc[100]
```

```python
# 데이터를 날짜 역순으로 정렬
df.sort_index(ascending=False)
```

```python
# 'Close' 데이터를 가격이 높은 순서대로 정렬
df['Close'].sort_values(ascending=False)
```

```python
# 인덱스 컬럼을 0부터 순차적으로 증가하는 정수로 변경
df.reset_index(inplace=True)
df
```

```python
# 컬럼명 변경
df.rename(columns = {'Date': '날짜', 'High': '고가', 'Low': '저가'})
```

```python
# 날짜 컬럼 삭제
df.drop(['Date'], axis=1)
df
```

```python
# 'diff' 컬럼 생성, 'Close'와 'Open'의 차이 계산하여 채우기
df['diff'] = df['Close'] - df['Open']
df
```

```python
# 'Date' 필드에 저장된 값을 파싱하여 'Year', 'Month', 'Day' 필드에 저장
df['Year'] = [d.year for d in df['Date']]
df['Month'] = [d.month for d in df['Date']]
df['Day'] = [d.day for d in df['Date']]
df
```

```python
# 모든 데이터를 소수점 둘째자리까지 표시
df.round(2)
```

```python
# 거래량 필드를 생성, 세 자리수마다 콤마가 들어가도록 변경
df['거래량'] = df['Volume'].apply(lambda x: "{:,}".format(x))
df
```

```python
# 주가 데이터를 'Month' 별로 그룹핑, 'Volume'의 평균값 표시
df['Volume'].groupby(df['Month']).mean()
```

```python
# 결측치 있는지 확인, 결측치 처리
df.isna().sum()
```

```python
# 12월의 데이터와 11월의 데이터를 별도의 데이터프레임으로 만든 후, 이어붙이기
dec_df = df[df['Month'] == 12]
nov_df = df[df['Month'] == 11]
nov_dec_df = pd.concat([nov_df, dec_df])
nov_dec_df
```
