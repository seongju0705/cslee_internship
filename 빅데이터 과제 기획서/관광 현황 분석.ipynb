## 2019년도와 2020년도 지역별 방문자 수 데이터 이용

```python
import pandas as pd
import numpy as np
```

```python
df2019 = pd.read_csv('2019 지역별 방문자 수.csv', encoding='euc-kr')
df2019
```

```python
df2020 = pd.read_csv('2020 지역별 방문자 수.csv', encoding='euc-kr')
df2020
```

```python
df = pd.merge(df2019, df2020, how='outer', on='광역지자체')
df
```

```python
df.rename(columns = {'방문자 수_x': '2019', '방문자 수_y': '2020'}, inplace = True)
df
```

```python
df['diff'] = df['2019'] - df['2020']
df
```

```python
df = df.sort_values(by=df.columns[3], ascending=False)
df
```
