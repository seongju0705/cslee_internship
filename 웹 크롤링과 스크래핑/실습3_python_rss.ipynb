```python
# ElementTree 모듈 로드
from xml.etree import ElementTree
```

```python
# parse() 함수로 파일을 읽고 ElementTree 객체를 생성
tree = ElementTree.parse('rss.xml')
```

```python
# getroot() 메서드로 XML의 루트 element를 추출
root = tree.getroot()
```

```python
import pandas as pd
```

```python
df_list = []
for item in root.findall('channel/item/description/body/location/data'):
    # find() 메서드로 element 탐색, text 속성으로 값을 추출
    tm_ef = item.find('tmEf').text
    tmn = item.find('tmn').text
    tmx = item.find('tmx').text
    wf = item.find('wf').text
    df = pd.DataFrame({
        '일시': [tm_ef],
        '최저기온': [tmn],
        '최고기온': [tmx],
        '날씨': [wf]
    })
    df_list.append(df)
weather = pd.concat(df_list)
weather
```

```python
type(weather)
```

```python
weather.to_csv('weather.csv')
```

```python
ex = pd.ExcelWriter('weather.xlsx')
weather.to_excel(ex, '.', index=False)
ex.save()
```

```python
weather.reset_index(drop=True, inplace=True)
```

```python
weather.to_json('weather.json')
```
